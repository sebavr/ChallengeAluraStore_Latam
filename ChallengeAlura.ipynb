# **Introducción**

El Sr. Juan busca iniciar su emprendimiento, por ello nuestro objetivo será el mostrarle cual es la mejor alternativa en base a ventas, rendimiento y reseñas.
Esta muestra de datos se hará por medio de gráficos y analisis de las tablas de datos de cada tienda con el fin de discriminar y seleccionar la mejor.






# **1) Importación de datos**

```python
dato1="https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
dato2="https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
dato3="https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
dato4="https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

import pandas as pd

tienda1 = pd.read_csv(dato1)
tienda2 = pd.read_csv(dato2)
tienda3 = pd.read_csv(dato3)
tienda4 = pd.read_csv(dato4)

tienda1.head()
```

# **2) Total por tienda**

```python
ingreso_tienda1 = tienda1["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

print(f'El ingreso de tienda 1 en millones de pesos  : {float(ingreso_tienda1/1000000):.2f}')
print(f'El ingreso de tienda 2  en millones de pesos es : {float(ingreso_tienda2/1000000):.2f}')
print(f'El ingreso de tienda 3 en millones de pesos es : {float(ingreso_tienda3/1000000):.2f}')
print(f'El ingreso de tienda 4 e en millones de pesos es : {float(ingreso_tienda4/1000000):.2f}')
```

Se puede ver que el mayor ingreso por tienda es para la tienda1, seguido por la tienda 2, tienda 3 y tienda 4.

# **4) Ingreso por categoría**

```python
import matplotlib.pyplot as plt


#Tienda 1
ingresos_por_categoria_tienda1 = tienda1.groupby('Categoría del Producto')['Precio'].sum()
print(f'El ingreso de Tienda 1, {ingresos_por_categoria_tienda1}')
# Ordena los ingresos por categoría en forma descendente
ingresos_por_categoria_tienda1 = ingresos_por_categoria_tienda1.sort_values(ascending=False)
ingresos_por_categoria_tienda1_mm = round(ingresos_por_categoria_tienda1/1000000,2)
ax = ingresos_por_categoria_tienda1_mm.plot(kind='bar', color = 'blue')
plt.bar_label(ax.containers[0])
plt.title('Ingreso por categoria de producto Tienda 1')
ax.set_xlabel('Categoría del Producto')
ax.set_ylabel('Ingreso en millones de pesos')
plt.show()

#Tienda 2
ingresos_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum()
print(f'El ingreso de Tienda 2, {ingresos_por_categoria_tienda2}')
# Ordena los ingresos por categoría en forma descendente
ingresos_por_categoria_tienda2 = ingresos_por_categoria_tienda2.sort_values(ascending=False)
ingresos_por_categoria_tienda2_mm = round(ingresos_por_categoria_tienda2/1000000,2)
ax = ingresos_por_categoria_tienda2_mm.plot(kind='bar', color = 'green')
plt.bar_label(ax.containers[0])
plt.title('Ingreso por categoria de producto Tienda 2')
ax.set_xlabel('Categoría del Producto')
ax.set_ylabel('Ingreso en millones de pesos')
plt.show()

#Tienda 3
ingresos_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum()
print(f'El ingreso de Tienda 3, {ingresos_por_categoria_tienda3}')
# Ordena los ingresos por categoría en forma descendente
ingresos_por_categoria_tienda3 = ingresos_por_categoria_tienda3.sort_values(ascending=False)
ingresos_por_categoria_tienda3_mm = round(ingresos_por_categoria_tienda3/1000000,2)
ax = ingresos_por_categoria_tienda3_mm.plot(kind='bar', color = 'red')
plt.bar_label(ax.containers[0])
plt.title('Ingreso por categoria de producto Tienda 3')
ax.set_xlabel('Categoría del Producto')
ax.set_ylabel('Ingreso en millones de pesos')
plt.show()

#Tienda 4
ingresos_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum()
print(f'El ingreso de Tienda 4, {ingresos_por_categoria_tienda4}')
# Ordena los ingresos por categoría en forma descendente
ingresos_por_categoria_tienda4 = ingresos_por_categoria_tienda4.sort_values(ascending=False)
ingresos_por_categoria_tienda4_mm = round(ingresos_por_categoria_tienda4/1000000,2)
ax = ingresos_por_categoria_tienda4_mm.plot(kind='bar', color = 'orange')
plt.bar_label(ax.containers[0])
plt.title('Ingreso por categoria de producto Tienda 4')
ax.set_xlabel('Categoría del Producto')
ax.set_ylabel('Ingreso en millones de pesos')
plt.show()
```

# **5) Valoración media por tienda**

```python
# Se obtiene la calificacion promedio
calficacion_promedio_tienda1 = float(tienda1["Calificación"].mean())
print(f'La calificacion promedio de tienda1 es: {calficacion_promedio_tienda1:.2f}')

# Se obtiene la calificacion promedio
calficacion_promedio_tienda2 = float(tienda2["Calificación"].mean())
print(f'La calificacion promedio de tienda2 es: {calficacion_promedio_tienda2:.2f}')

# Se obtiene la calificacion promedio
calficacion_promedio_tienda3 = float(tienda3["Calificación"].mean())
print(f'La calificacion promedio de tienda3 es: {calficacion_promedio_tienda3:.2f}')

# Se obtiene la calificacion promedio
calficacion_promedio_tienda4 = float(tienda4["Calificación"].mean())
print(f'La calificacion promedio de tienda4 es: {calficacion_promedio_tienda4:.2f}')
```
